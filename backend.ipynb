```python
from googletrans import Translator, LANGUAGES

def translate_text_google(text, src_lang='en', dest_lang='de'):
    translator = Translator()
    translation = translator.translate(text, src=src_lang, dest=dest_lang)
    return translation.text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity(model_sentence, google_translate_sentence):
    vectorizer = TfidfVectorizer().fit([model_sentence, google_translate_sentence])
    vectorized_sentences = vectorizer.transform([model_sentence, google_translate_sentence])
    similarity = cosine_similarity(vectorized_sentences[0:1], vectorized_sentences[1:2])
    return similarity[0][0]
```
